```python
# Run Census Search to retrieve data on all states and counties
r = requests.get(f'https://api.census.gov/data/2018/acs/acs5?get=NAME,B01001_001E,B01002_001E,B19013_001E,B19301_001E,B17001_002E,B23025_005E,B02001_002E,B02001_003E,B03003_003E,B05001_002E,B05001_005E&for=county:*&in=state:*&key={api_key}')
r.status_code
data = r.json()

# Convert to DataFrame
df = pd.DataFrame(data)
header = df.iloc[0]
df = df[1:]
df.columns = header

# Column Re-naming
census_pd = df.rename(columns = {'NAME': 'county&state',
                "B01001_001E": "Population",
                "B01002_001E": "Median Age",
                "B19013_001E": "Household Income",
                "B19301_001E": "Per Capita Income",
                "B17001_002E": "Poverty Count",
                "B23025_005E": "Unemployment Count",
                "B02001_002E" : "Race: White",              
                "B02001_003E" : "Race: African American",
                "B03003_003E" : "Race: Hispanic or Latino",
                "B05001_002E": "US Citizen, born in US",
                "B05001_005E": "US Citizen, by naturalization"})
```
